# Interact with website

```python
import mechanicalsoup
```

```python
browser = mechanicalsoup.StatefulBrowser()
browser.open(url)
page = browser.get("https://github.com/login")

print('We are in!')
```

```python
browser.select_form('form[action="./ConsultarBMPAberto.aspx"]')
```

```python
browser["ctl00$Conteudo$ddlEmpresas"] = "Boa Vista Roraima Energia S.A." #set empresa
browser["ctl00$Conteudo$txtDataInicioPesquisa"] = "012020"
browser["ctl00$Conteudo$txtDataFimPesquisa"] = "012020"
browser["ctl00$Conteudo$btnConfirmar"] = "Confirmar"
```

```python
resp =  browser.submit_selected()
```

```python
print(resp.text)
```

```python
print(resp.soup)
```
